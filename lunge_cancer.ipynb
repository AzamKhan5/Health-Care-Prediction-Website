```python
y = df['LUNG_CANCER']
x = df.drop('LUNG_CANCER', axis = 1)
```

```python
# Split the data into trian and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, random_state=25, stratify= y)
```

```python
print("x_train",x_train.shape)
print("x_test",x_test.shape)
```

```python
y_train.shape
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train
```

```python
x_train_sc = sc.fit_transform(x_train_smote)
x_test_sc = sc.transform(x_test)
```

```python
x_train_sc
```

```python
sc.inverse_transform([x_train_sc[0]])
```

```python
x_test_sc
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=42)
model.fit(x_train_sc, y_train_smote)
```

```python
y_pred = model.predict(x_test_sc)
len(y_pred)
```

```python
src = pd.Series(y_pred)
src.value_counts(normalize = True)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
model.score(x_train_sc,y_train_smote)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, fmt = '.2f')
```

```python
import pickle

with open('lung_cancer_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
with open('lung_cancer_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)


y_pred_loaded = loaded_model.predict(x_test_sc)

print("Predictions from loaded model:", y_pred_loaded)
```
