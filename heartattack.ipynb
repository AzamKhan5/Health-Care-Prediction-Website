```python
x = f.drop('target',axis=1)
x.head()
```

```python
y = f['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=180, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train[:5]
```

```python
x_test[:5]
```

```python
# check the shape of splited data
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.svm import SVC
sv = SVC()
```

```python
sv.fit(x_train, y_train)
```

```python
sv.score(x_train, y_train)
```

```python
ypred = sv.predict(x_test)
ypred[:5]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

accuracy_score(y_test , ypred)
```

```python
print(classification_report(y_test ,ypred))
```

```python
confusion_matrix(y_test , ypred)
```

```python
 sns.heatmap(confusion_matrix(y_test, ypred), annot = True, fmt = '.2f')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
new_sv = SVC()
```

```python
para_grid = {'C':np.arange(1,10,0.5),
    'kernel':['rbf','poly','linear'],
    'gamma':['scale', 'auto']}
```

```python
new_sv = GridSearchCV(new_sv, para_grid, refit = True, verbose = 2)
```

```python
new_sv.fit(x_train, y_train)
```

```python
new_sv.best_params_
```

```python
new_sv.best_score_
```

```python
model  = SVC(C=2.5, gamma='scale', kernel='poly')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model_ypred =model.predict(x_test)
```

```python
accuracy_score(y_test, model_ypred)
```
